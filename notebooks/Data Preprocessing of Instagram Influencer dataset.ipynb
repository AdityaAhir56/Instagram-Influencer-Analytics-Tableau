## **About the dataset (Instagram Influencer dataset)**
The data set contains 200 observations. Information about the columns are given below.

#### Data Dictionary
- **`rank`**: This column represents the **rank** of that particular user.
  
- **`channel_info`**: This column represents the Username of that user.
- **`influence_score`**: Influence score of the users. It is calculated on basis of mentions, importance and popularity.
- **`posts`**: Number of posts they have made so far.
- **`followers`**: Number of followers of the user.
- **`avg_likes`**: Average likes on User's posts (total likes/ total posts).
- **`60_day_eng_rate`**: Last 60 days engagement rate of instagrammer as faction of engagements they have done so far.
- **`new_post_avg_like`**: Average likes they have on new posts.
- **`total_likes`**: Total likes the user has got on their posts (in Billion).
- **`country`**: Country or region of origin of the user.

**Interpretation**

From this we can see that:
- There are **2 Numerical columns**.
- And there are **10 Object** type columns.

# Step-2: Data Preprocessing

## a) Renaming the columns

```python
# Renaming the necessary columns appropriately
df_influencer_data = df_influencer_data.rename(columns = {"rank": "Rank_of_User",
                                                          "channel_info": "Username",
                                                          "influence_score": "Influence_Score",
                                                          "posts": "No_of_Posts",
                                                          "followers": "No_of_Followers",
                                                          "avg_likes": "Avg_No_of_Likes",
                                                          "60_day_eng_rate": "Engagement_rate_of_last_60_days_in_Percentage",
                                                          "new_post_avg_like": "Avg_No_of_likes_on_new_posts",
                                                          "total_likes": "Total_No_of_likes",
                                                          "country": "Country_of_User"})
```

## b) Correcting the datatypes and values according to the datatypes of the columns

```python
df_influencer_data.info()
```

```python
df_influencer_data.head()
```

```python
# Correcting values and datatype of "No_of_Posts" column

# Correcting values
for i in df_influencer_data['No_of_Posts']:
    index = df_influencer_data[df_influencer_data['No_of_Posts'] == i].index[0]
    if i.startswith('0'):
        i = i.removeprefix('0')
        if i.find('.') == 0:
            i = i.removeprefix('.')
            if i.startswith('0'):
                i = i.removeprefix('0')
                i = i.replace('k', '0')
                df_influencer_data['No_of_Posts'][index] = i
            else:
                i = i.replace('k', '0')
                df_influencer_data['No_of_Posts'][index] = i
    elif i.find('.'):
        i = i.replace('.', '')
        i = i.replace('k', '00')
        df_influencer_data['No_of_Posts'][index] = i

# Correcting datatype
df_influencer_data['No_of_Posts'] = df_influencer_data['No_of_Posts'].astype('int64')
```

```python
# Correcting values and datatype of "No_of_Followers" column

# Correcting values
for i in df_influencer_data['No_of_Followers']:
    index = df_influencer_data[df_influencer_data['No_of_Followers'] == i].index[0]
    if i.endswith('m'):
        i = i.replace('.', '')
        i = i.replace('m', '00000')
        df_influencer_data['No_of_Followers'][index] = i
    else:
        i = i.replace('.', '')
        df_influencer_data['No_of_Followers'][index] = i

# Correcting datatype
df_influencer_data['No_of_Followers'] = df_influencer_data['No_of_Followers'].astype('int64')
```

```python
# Correcting values and datatype of "Avg_No_of_Likes" column

# Correcting values
for i in df_influencer_data['Avg_No_of_Likes']:
    index = df_influencer_data[df_influencer_data['Avg_No_of_Likes'] == i].index[0]
    if i.endswith('k'):
        i = i.replace('k', '00')
        if i.find('.'):
            i = i.replace('.', '')
            df_influencer_data['Avg_No_of_Likes'][index] = i
    elif i.endswith('m'):
        i = i.replace('m', '00000')
        if i.find('.'):
            i = i.replace('.', '')
            df_influencer_data['Avg_No_of_Likes'][index] = i

# Correcting datatype
df_influencer_data['Avg_No_of_Likes'] = df_influencer_data['Avg_No_of_Likes'].astype('int64')
```

```python
# Correcting values and datatype of "Engagement_rate_of_last_60_days_in_Percentage" column

# Correcting values
for i in df_influencer_data['Engagement_rate_of_last_60_days_in_Percentage']:
    index = df_influencer_data[df_influencer_data['Engagement_rate_of_last_60_days_in_Percentage'] == i].index[0]
    if i.endswith('%'):
        i = i.replace('%', '')
        df_influencer_data['Engagement_rate_of_last_60_days_in_Percentage'][index] = i

# Correcting datatype
df_influencer_data['Engagement_rate_of_last_60_days_in_Percentage'] = df_influencer_data['Engagement_rate_of_last_60_days_in_Percentage'].astype('float64')
```

```python
# Correcting values and datatype of "Avg_No_of_likes_on_new_posts" column

# Correcting values
for i in df_influencer_data['Avg_No_of_likes_on_new_posts']:
    index = df_influencer_data[df_influencer_data['Avg_No_of_likes_on_new_posts'] == i].index[0]
    if i.endswith('k'):
        i = i.replace('k', '00')
        if i.find('.'):
            i = i.replace('.', '')
            df_influencer_data['Avg_No_of_likes_on_new_posts'][index] = i
    elif i.endswith('m'):
        i = i.replace('m', '00000')
        if i.find('.'):
            i = i.replace('.', '')
            df_influencer_data['Avg_No_of_likes_on_new_posts'][index] = i

# Correcting datatype
df_influencer_data['Avg_No_of_likes_on_new_posts'] = df_influencer_data['Avg_No_of_likes_on_new_posts'].astype('int64')
```

```python
# Correcting values and datatype of "Total_No_of_likes" column

# Correcting values
for i in df_influencer_data['Total_No_of_likes']:
    index = df_influencer_data[df_influencer_data['Total_No_of_likes'] == i].index[0]
    if i.endswith('m'):
        i = i.replace('m', '00000')
        if i.find('.'):
            i = i.replace('.', '')
            df_influencer_data['Total_No_of_likes'][index] = i
    elif i.endswith('b'):
        i = i.replace('b', '00000000')
        if i.find('.'):
            i = i.replace('.', '')
            df_influencer_data['Total_No_of_likes'][index] = i

# Correcting datatype
df_influencer_data['Total_No_of_likes'] = df_influencer_data['Total_No_of_likes'].astype('int64')
```

## c) Null value Handling

```python
# Checking how many null values are present in the dataset
df_influencer_data.isnull().sum()
```

**Interpretation**

Here we can see that there are only two column with any null values **`Engagement_rate_of_last_60_days_Percentage`**, **`Country_of_User`**. Now before doing null value treatment we'll have to see these null values in percentage to do the null value treatment accordingly.

```python
# Checking how many null values are present in the dataset in percentage
((df_influencer_data.isna().sum())*100)/len(df_influencer_data)
```

**Interpretation**

From the null value analysis, we can see that there are **31%** missing values in the **`Country_of_User`** column, now we will apply the following rule:

**Rule Applied**:
- If null values < **15%** → Dropping the null values.  
- If null values between **15–70%** → Impute (mean/median/mode).  
- If null values > **70%** → Drop column.  

---

**Decision**:  
- Since the column has **31% null values which is between 15–70%**, so we will **impute the missing values** and though it is a categorical column but we cannot impute using **mode** because it is small dataset and the column we are dealing with is **`Country`** which is not easy to impute, so we will impute it with a constant **Unknown**.
- And for **`Engagement_rate_of_last_60_days_Percentage`** column we will just do **dropna**.

```python
# Imputing the column with it's mean
df_influencer_data.fillna({"Country_of_User": "Unknown"}, inplace = True)
```

```python
# Dropping rest of the null values
df_influencer_data.dropna(inplace = True)
```
